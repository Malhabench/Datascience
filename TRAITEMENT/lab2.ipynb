```python
#returns the image matrix

import cv2
import numpy as np
import matplotlib.pyplot as plt

def OpenImage(image_path):
    I = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Lecture en niveaux de gris
    L, C = I.shape
    return I, L, C
```

```python
# Split the image into its three color channels (Red, Green, Blue)

def Divide(image_path):
    I = cv2.imread(image_path)
    B, G, R = cv2.split(I)
    return B, G, R
```

```python
#converts the image from BGR to HSV format

def HSV(image_path):
    I = cv2.imread(image_path)
    hsv = cv2.cvtColor(I, cv2.COLOR_BGR2HSV)
    return hsv
```

```python
#calculate the total number of pixels in the image

def CountPix(image_path):
    I = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    return I.size
```

```python
#modifies the intensity of the image by multiplying it by a factor.


def FactPix(image_path, alpha, beta):
    I = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    I_fact = cv2.convertScaleAbs(I, alpha=alpha, beta=beta)
    return I_fact
```

```python
# computes the logarithm, exponential, square, and square root of the image

def Func_a(image_path):
    I = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    log = np.log1p(I)
    exp = np.exp(I/255.0)
    square = np.square(I)
    sqrt = np.sqrt(I)
    return log, exp, square, sqrt
```

```python
#calculate the mean and standard deviation

def Func_m(image_path):
    I = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    mean, stddev = cv2.meanStdDev(I)
    return mean[0][0], stddev[0][0]
```

```python
#normalizes the image

def Normalize(image_path, min_val, max_val):
    I = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    Inorm = cv2.normalize(I, None, min_val, max_val, cv2.NORM_MINMAX)
    return Inorm
```

```python
#compute the inverse 

def Inverse(image_path):
    I = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    Inv = 255 - I
    return Inv
```

```python
#calculate the histogram of the image.

def CalcHist(image_path):
    I = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    H, bins = np.histogram(I, bins=256, range=[0, 256])
    return H, bins
```

```python
#applies a threshold to the image

def Threshold(image_path, s):
    I = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    _, T = cv2.threshold(I, s, 255, cv2.THRESH_BINARY)
    return T
```

```python
import matplotlib.pyplot as plt

def Func_j(image_path):
    I, _, _ = OpenImage(image_path)
    plt.figure(figsize=(12, 6))
    plt.subplot(231), plt.imshow(I, cmap='gray'), plt.title('Original')
    H, bins = CalcHist(image_path)
    plt.subplot(232), plt.plot(H), plt.title('Histogram')
    Inv = Inverse(image_path) 
    H_inv, bins_inv = np.histogram(Inv, bins=256, range=[0, 256]) 
    plt.subplot(233), plt.imshow(Inv, cmap='gray'), plt.title('Inverse')
    plt.subplot(234), plt.plot(bins_inv[:-1], H_inv), plt.title('Histogram Inverse')
    plt.show()
```

```python


def Func_t(image_path):
    I, _, _ = OpenImage(image_path)
    Inorm = Normalize(image_path, 10, 50)
    plt.figure(figsize=(12, 6))
    plt.subplot(231), plt.imshow(I, cmap='gray'), plt.title('Original')
    H, bins = CalcHist(image_path)
    plt.subplot(232), plt.plot(bins[:-1], H), plt.title('Histogram')
    plt.subplot(233), plt.imshow(Inorm, cmap='gray'), plt.title('Normalized')
    H_norm, bins_norm = np.histogram(Inorm, bins=256, range=[0, 256])
    plt.subplot(234), plt.plot(bins_norm[:-1], H_norm), plt.title('Histogram Normalized')
    plt.show()
```

```python


def Func_f(image_path):
    I, _, _ = OpenImage(image_path)
    T = Threshold(image_path, 128)
    plt.figure(figsize=(12, 6))
    plt.subplot(231), plt.imshow(I, cmap='gray'), plt.title('Original')
    H, bins = CalcHist(image_path)
    plt.subplot(232), plt.plot(bins[:-1], H), plt.title('Histogram')
    plt.subplot(233), plt.imshow(T, cmap='gray'), plt.title('Thresholded')
    H_th, bins_th = np.histogram(T, bins=256, range=[0, 256])
    plt.subplot(234), plt.plot(bins_th[:-1], H_th), plt.title('Histogram Thresholded')
    plt.show()
```
